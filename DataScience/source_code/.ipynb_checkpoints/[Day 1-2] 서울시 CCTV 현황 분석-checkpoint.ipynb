```python
import pandas as pd
cctv_seoul = pd.read_excel('../data/01. cctv_in_seoul.xlsx',
                          encoding='utf-8')
cctv_seoul.head()
```

```python
pop_seoul = pd.read_excel('../data/01. population_in_seoul.xls',
                          header=2,
                          usecols='B, D, G, J, N',
                         encoding='utf-8')
pop_seoul.head()
```

#### QUIZ! 외국인 비율, 고령자 비율 계산
- 외국인 비율이 가장 높은 곳? : 영등포
- 외국인 비율이 가장 낮은 곳? : 도봉구
- 고령자 비율이 가장 높은 곳? : 강북구
- 고령자 비율이 가장 낮은 곳? : 송파구

```python
pop_seoul['외국인비율'] = pop_seoul['외국인'] / pop_seoul['인구수'] * 100
pop_seoul['고령자비율'] = pop_seoul['고령자'] / pop_seoul['인구수'] * 100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by=["인구수", "고령자비율"], ascending=False).head()
```

## CCTV + 인구수 병합! (p.53)

```python
pop_seoul.head()
```

```python
cctv_seoul.head()
```

```python
cctv_seoul['구별'] = cctv_seoul['구별'].replace(
    to_replace=" ",  # 어떤 string을?
    value="", # 무엇으로 ?
    regex = True # regex 유무
)
cctv_seoul.head()
```

```python
data_result = pd.merge(cctv_seoul, pop_seoul, on="구별")
data_result.head()
```

```python
del data_result['2011년 이전']
del data_result['2012년']
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

#### (cf) 피어슨 상관계수 (corrcoef)
- 절대값 0.1 이하면 무시
- 절대값 0.3 이하면 `약한 상관관계`
- 절대값 0.7 이하면 `뚜렷한 상관관계`

```python
import numpy as np
np.corrcoef(data_result['고령자비율'], data_result['소계'])
```

```python
np.corrcoef(data_result['인구수'], data_result['소계'])
```

```python
np.corrcoef(data_result['한국인'], data_result['소계'])
```

```python
np.corrcoef(data_result['외국인비율'], data_result['소계'])
```
