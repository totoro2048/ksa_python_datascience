```python
station_name = []

for name in crime_anal_police['관서명']:
    station_name.append('서울'+name[:-1]+'경찰서')
```

```python
station_address = [] 
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language='ko')
    address = tmp[0].get('formatted_address')
    
    station_address.append(address)
    
    tmp_loc = tmp[0].get('geometry')
    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name+'-->'+address)
```

```python
gu_name = []

for name in station_address:
    tmp = name.split() # 빈칸 기준으로 string 자르기
    # 마지막 글자가 `구`인 것 중에 첫번째 것
    tmp_gu = [gu for gu in tmp if gu[-1] == '구'][0]
    gu_name.append(tmp_gu)

# for name in station_address:
#     tmp = name.split()
#     for gu in tmp:
#         if gu[-1] == '구':
#             tmp_gu = gu
#     gu_name.append(tmp_gu)

crime_anal_police['구별'] = gu_name
crime_anal_police.head()
```

## pivot_table in pandas
- 중복된 row 합치기 !!!!!!!!!!!

```python
df = pd.read_excel('../data/02. sales-funnel.xlsx')
df.head()
```

```python
pd.pivot_table(df, index=['Name']) 
# 별도 설정 없으면 숫자 column만 남는다
# value는 평균값
```

```python
pd.pivot_table(df, index=['Name', 'Rep', 'Manager'])
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'])
```

```python
pd.pivot_table(df, 
               index=['Manager', 'Rep'], 
               values=['Price'],
               aggfunc=np.sum) # 합계로 표시
```

```python
pd.pivot_table(df,
              index=['Manager', 'Rep'],
              values=['Price'],
              columns=['Product'],
              aggfunc=[np.sum],
              fill_value=0)
```

```python
pd.pivot_table(df,
              index=['Manager', 'Rep','Product'],
              values=['Price', 'Quantity'],
              aggfunc=[np.sum, np.mean],
              fill_value=0,
              margins=True)
```
